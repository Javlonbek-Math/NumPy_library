DataFrame data structure

```python
from pandas import DataFrame
```

```python
data = {'Yil':[2021, 2020, 2019, 2018, 2017, 2016, 2015, 2010],
       'Aholi soni':[33.9, 33.5, 32.9, 32.5, 31.9, 31.4, 30.9, 28.5],
       "Temp":[1.54, 1.48, 1.56, 1.62, 1.65, 1.66, 1.64, 1.53]}
df1 = DataFrame(data)
df1
```

```python
df1=DataFrame(data, columns=['Yil','Temp','Aholi soni','YAIM'])
df1
```

```python
df1 = DataFrame(data, columns=['Yil','Temp','Aholi soni','YAIM'], index=[21, 20, 19, 18, 17, 16, 15, 10])
df1
```

```python
df1.YAIM = 1600
```

```python
df1
```
